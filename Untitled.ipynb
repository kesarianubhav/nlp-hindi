```python
def tokenize_file(filename):
    file=[]
    gram_file=[]
    with open(filename,'r') as f:
        for line in f:
            for word in line.split():
                file.append(word)
    return file

def create_k_grams(filename , gram_size):
    gram_file=[]
    file=tokenize_file(filename)
    size=len(file)
    for i in range(0,size,1):
        p=file[i:i+int(gram_size)]
        gram_file.append(p)
        i=i+gram_size+1        
#     print(gram_file)
    return gram_file
    
```

```python
def jaccard_similarity(a, b):
    intersection = set(a).intersection(set(b))
    union = set(a).union(set(b))
    return len(a)/len(b)
    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def top_words(text):
#     print(te)
    vectorizer = TfidfVectorizer()
    matrix = vectorizer.fit_transform(text).todense()
    matrix = pd.DataFrame(matrix, columns=vectorizer.get_feature_names())
    top_words = matrix.sum(axis=0).sort_values(ascending=False)
    return top_words
```

```python
text = tokenize_file('headlines.txt')
k_grams=create_k_grams('headlines.txt',5)
# print(k_grams)
top_words=top_words(text)
print(top_words)
```

```python
k_grams1=create_k_grams('headlines.txt',4)
k_grams2=create_k_grams('headlines2.txt',4)
k_grams1 = set(tuple(x) for x in k_grams1)
k_grams2 = set(tuple(x) for x in k_grams2)
print(jaccard_similarity(k_grams2,k_grams1))
```
